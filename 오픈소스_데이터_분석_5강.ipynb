## 5-2 JSON 형식 저장

```python
### 라이브러리 임포트
import json
import pandas as pd

data = {
    "이름": "홍길동",
    "나이": 25,
    "거주지": "서울",
    "관심사": ["프로그래밍", "데이터 분석", "여행"]
}

## json.dump를 이용한 저장, 저장모드는 write즉 기록하는모드
# json.dump라는 함수로 데이터저장가능,
# data를 f에 저장하겠다, 인덴트 옵션 넣지않으면 데이터계층구분어렵다, json은 기본적으로 한글저장안하고 아스키코드만 됨
# 따라서 ascii=False로 옵션을 주어야 한글로 저장이 됨
with open('output.json', mode='w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

# 왼쪽 파일란에 output.json 파일이 생성된 것을 확인할 수 있다


## DataFrame을 이용한 저장
# 앞서 ensure_ascii와 아래의 force_ascii는 같은 목적이나 이름이 다름
# orient옵션에 따라(records, column 등) 출력방식이 달라짐, 그전에 리스트형태로 변경[]
df = pd.DataFrame(data)
df.to_json('output_df.json', orient='records', indent=4, force_ascii=False)
```
